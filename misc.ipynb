```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
balance_range = np.arange(-1, 1.01, 0.01)
```

```python
balance_weight = np.cos(np.pi / 2 * balance_range)
```

```python
balance_weight = np.cos(np.pi / 3 * balance_range)
```

```python
balance_weight = 1.2* np.power(np.cos(np.pi / 3 * balance_range), 0.8)
```

```python
plt.plot(balance_range, balance_weight)
plt.ylim(0, 1.2)
```

```python
contrast_range = np.arange(0, 256, 0.1)
```

```python
contrast_weight = contrast_range/32
contrast_weight = contrast_weight.clip(0, 1)
```

```python
contrast_weight = np.power(contrast_range/32, 0.2)
contrast_weight = contrast_weight.clip(0, 1.2)
```

```python
plt.plot(contrast_range, contrast_weight)
```
